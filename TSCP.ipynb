### Model Architecture

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.
import tensorflow as tf
EPOCHS = 1
BATCH_SIZE = 128
````

````python
from tensorflow.contrib.layers import flatten

def LeNet(x0,x1,x2):
    mu = 0
    sigma = 0.1
    
    W_conv1 = tf.Variable(tf.truncated_normal(shape=(5,5,1,6), mean = mu, stddev = sigma)) 
    b_conv1 = tf.Variable(tf.zeros(6))
    conv1 = tf.nn.conv2d(x0, W_conv1, strides = [1,1,1,1], padding = 'VALID') + b_conv1
    
    conv1 = tf.nn.relu(conv1)
    
    conv1 = tf.nn.max_pool(conv1, ksize = [1,2,2,1], strides = [1,2,2,1], padding = 'VALID')
    
    W_conv2 = tf.Variable(tf.truncated_normal(shape=(5,5,6,16), mean = mu, stddev = sigma))
    b_conv2 = tf.Variable(tf.zeros(16))
    conv2 = tf.nn.conv2d(conv1, W_conv2, strides=[1,1,1,1], padding='VALID') + b_conv2
    
    conv2 = tf.nn.relu(conv2)
    
    conv2 = tf.nn.max_pool(conv2, ksize=[1,2,2,1], strides = [1,2,2,1], padding= 'VALID')
    
    fc0 = flatten(conv2)
    
    W_fc1 = tf.Variable(tf.truncated_normal(shape=(400,120), mean = mu, stddev = sigma))
    b_fc1 = tf.Variable(tf.zeros(120))    
    fc1 = tf.add(tf.matmul(fc0, W_fc1), b_fc1)
    
    fc1 = tf.nn.relu(fc1)
    
    W_fc2 = tf.Variable(tf.truncated_normal(shape=(120,84), mean = mu, stddev = sigma))
    b_fc2 = tf.Variable(tf.zeros(84))
    fc2 = tf.add(tf.matmul(fc1, W_fc2), b_fc2)
    
    fc2 = tf.nn.relu(fc2)
    
    W_fc3 = tf.Variable(tf.truncated_normal(shape=(84,43), mean = mu, stddev = sigma))
    b_fc3 = tf.Variable(tf.zeros(43))
    logits1 = tf.add(tf.matmul(fc2, W_fc3), b_fc3)
    
    
    W_conv1 = tf.Variable(tf.truncated_normal(shape=(5,5,1,6), mean = mu, stddev = sigma)) 
    b_conv1 = tf.Variable(tf.zeros(6))
    conv1 = tf.nn.conv2d(x1, W_conv1, strides = [1,1,1,1], padding = 'VALID') + b_conv1
    
    conv1 = tf.nn.relu(conv1)
    
    conv1 = tf.nn.max_pool(conv1, ksize = [1,2,2,1], strides = [1,2,2,1], padding = 'VALID')
    
    W_conv2 = tf.Variable(tf.truncated_normal(shape=(5,5,6,16), mean = mu, stddev = sigma))
    b_conv2 = tf.Variable(tf.zeros(16))
    conv2 = tf.nn.conv2d(conv1, W_conv2, strides=[1,1,1,1], padding='VALID') + b_conv2
    
    conv2 = tf.nn.relu(conv2)
    
    conv2 = tf.nn.max_pool(conv2, ksize=[1,2,2,1], strides = [1,2,2,1], padding= 'VALID')
    
    fc0 = flatten(conv2)
    
    W_fc1 = tf.Variable(tf.truncated_normal(shape=(400,120), mean = mu, stddev = sigma))
    b_fc1 = tf.Variable(tf.zeros(120))    
    fc1 = tf.add(tf.matmul(fc0, W_fc1), b_fc1)
    
    fc1 = tf.nn.relu(fc1)
    
    W_fc2 = tf.Variable(tf.truncated_normal(shape=(120,84), mean = mu, stddev = sigma))
    b_fc2 = tf.Variable(tf.zeros(84))
    fc2 = tf.add(tf.matmul(fc1, W_fc2), b_fc2)
    
    fc2 = tf.nn.relu(fc2)
    
    W_fc3 = tf.Variable(tf.truncated_normal(shape=(84,43), mean = mu, stddev = sigma))
    b_fc3 = tf.Variable(tf.zeros(43))
    logits2 = tf.add(tf.matmul(fc2, W_fc3), b_fc3)
    
    W_conv1 = tf.Variable(tf.truncated_normal(shape=(5,5,1,6), mean = mu, stddev = sigma)) 
    b_conv1 = tf.Variable(tf.zeros(6))
    conv1 = tf.nn.conv2d(x2, W_conv1, strides = [1,1,1,1], padding = 'VALID') + b_conv1
    
    conv1 = tf.nn.relu(conv1)
    
    conv1 = tf.nn.max_pool(conv1, ksize = [1,2,2,1], strides = [1,2,2,1], padding = 'VALID')
    
    W_conv2 = tf.Variable(tf.truncated_normal(shape=(5,5,6,16), mean = mu, stddev = sigma))
    b_conv2 = tf.Variable(tf.zeros(16))
    conv2 = tf.nn.conv2d(conv1, W_conv2, strides=[1,1,1,1], padding='VALID') + b_conv2
    
    conv2 = tf.nn.relu(conv2)
    
    conv2 = tf.nn.max_pool(conv2, ksize=[1,2,2,1], strides = [1,2,2,1], padding= 'VALID')
    
    fc0 = flatten(conv2)
    
    W_fc1 = tf.Variable(tf.truncated_normal(shape=(400,120), mean = mu, stddev = sigma))
    b_fc1 = tf.Variable(tf.zeros(120))    
    fc1 = tf.add(tf.matmul(fc0, W_fc1), b_fc1)
    
    fc1 = tf.nn.relu(fc1)
    
    W_fc2 = tf.Variable(tf.truncated_normal(shape=(120,84), mean = mu, stddev = sigma))
    b_fc2 = tf.Variable(tf.zeros(84))
    fc2 = tf.add(tf.matmul(fc1, W_fc2), b_fc2)
    
    fc2 = tf.nn.relu(fc2)
    
    W_fc3 = tf.Variable(tf.truncated_normal(shape=(84,43), mean = mu, stddev = sigma))
    b_fc3 = tf.Variable(tf.zeros(43))
    logits3 = tf.add(tf.matmul(fc2, W_fc3), b_fc3)
    
    logits = (logits1+logits2+logits3)/3
    return logits
    
    
````

````python
x = tf.placeholder(tf.float32, (None, 32,32,3))
x0 = tf.placeholder(tf.float32, (None, 32,32,1))
x1 = tf.placeholder(tf.float32, (None, 32,32,1))
x2 = tf.placeholder(tf.float32, (None, 32,32,1))

y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, 43)
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the test set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.
rate = 0.0001
logits = LeNet(x0,x1,x2)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate=rate)
training_operation = optimizer.minimize(loss_operation)

````

````python
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        
        batch_x0 = batch_x[:,:,:,0]
        batch_x0 = batch_x0.reshape(batch_x.shape[0],32,32,1)

        batch_x1 = batch_x[:,:,:,1]
        batch_x1 = batch_x1.reshape(batch_x.shape[0],32,32,1)

        batch_x2 = batch_x[:,:,:,2]
        batch_x2 = batch_x2.reshape(batch_x.shape[0],32,32,1)
        
        
        
        accuracy = sess.run(accuracy_operation, feed_dict= {x0: batch_x0, x1: batch_x1, x2: batch_x2, y: batch_y})
        total_accuracy += (accuracy * len(batch_x))
        
    return total_accuracy / num_examples

````

````python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    print('Training')
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            
            batch_x0 = batch_x[:,:,:,0]
            batch_x0 = batch_x0.reshape(batch_x.shape[0],32,32,1)

            batch_x1 = batch_x[:,:,:,1]
            batch_x1 = batch_x1.reshape(batch_x.shape[0],32,32,1)

            batch_x2 = batch_x[:,:,:,2]
            batch_x2 = batch_x2.reshape(batch_x.shape[0],32,32,1)
            
            sess.run(training_operation, feed_dict={x0:batch_x0,x1:batch_x1,x2:batch_x2, y: batch_y})
        
        validation_accuracy = evaluate(X_valid, y_valid)
        print("EPOCH {} ... ".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
    saver.save(sess, './lenet')
    print("Model saved")    
````
